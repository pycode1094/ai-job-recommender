```python
import pandas as pd
```

```python
# 파일 경로에 맞게 수정 필요
df_trainee = pd.read_excel("./data/훈련생내역.xlsx")
df_preference = pd.read_excel("./data/취업희망내역(J형).xlsx")
```

```python
# 성/나이 컬럼 분리: "남/30" → ["남", "30"]
df_trainee[['성별', '나이']] = df_trainee['성/나이'].str.split('/', expand=True)

# 나이를 숫자형(int)으로 변환
df_trainee['나이'] = df_trainee['나이'].astype(int)

# 원래 컬럼 제거
df_trainee.drop(columns=['성/나이'], inplace=True)
```

```python
# '학번' 기준으로 병합 (훈련생 기준 전체 유지)
merged_df = pd.merge(df_trainee, df_preference, on='학번', how='left')
```

```python
merged_df.columns
```

```python
# 필요한 컬럼 지정
columns_trainee = [
    "학번", "이름_x", '성별', '나이', "생년월일", '핸드폰', "E-Mail", "훈련구분_x", "과정명_x", "훈련시작일", "훈련종료일"
]

columns_preference = [
    "희망여부", "(1지망)지역", "(1지망)직종", "(1지망)업종", "(1지망)보수", "(1지망)거주형태", "장래계획", "현당면문제점"
]

# 컬럼 추출 및 정제
df_selected = merged_df[columns_trainee + columns_preference].copy()

# 컬럼명 깔끔하게 정제
df_selected.columns = [
    "학번", "이름", '성별', '나이', "생년월일", '전화번호', "이메일", "훈련구분", "과정명", "훈련시작일", "훈련종료일",
    "희망여부", "희망지역", "희망직종", "희망업종", "희망보수", "희망거주형태", "장래계획", "당면문제"
]
```

```python
# 머지파일 엑셀뱉기
df_selected.to_excel('./data/merged_data.xlsx')
```
